<h4>Part 4(i): Creating the function to retrieve contents in a single tweet</h4>
<p>This function will take the XPATH of a single tweet as input, the function will then look at the single tweet input and identify the XPATHS of:
    <ol>
        <li>Tweet content - XPATH: ".//div[@data-testid="tweetText"]"</li>
        <li>Date/Time of tweet posted - XPATH: ".//time"</li>
        <li>number of replies - XPATH: ".//div[@data-testid = "reply"]"</li>
        <li>number of retweets - XPATH: ".//div[@data-testid = "retweet"]"</li>
        <li>number of likes - XPATH: ".//div[@data-testid = "like"]"</li>
    </ol>
    Using the XPATH, the function will return the text values of the contents, number of replies, likes and replies. For the time, the attribute 'datetime' will be returned. How this function will be used will be explained later.</p>


```python
#Function to extract a single tweet contents
def get_tweet_contents(card):
    
    try:
        #Tweet Date and time
        dateTime = card.find_element(By.XPATH, './/time').get_attribute('datetime')

        #Get content of text
        textContent = str(card.find_element(By.XPATH, './/div[@data-testid="tweetText"]').text)
    
    except NoSuchElementException:
        return
    #Get amount of likes, replies and retweets

    #Replies
    replies = card.find_element(By.XPATH, './/div[@data-testid = "reply"]').text
    #Retweets
    retweets = card.find_element(By.XPATH, './/div[@data-testid = "retweet"]').text
    #Likes
    likes = card.find_element(By.XPATH, './/div[@data-testid = "like"]').text
    
    #Tweet will contain textContent, dateTime, replies, retweets and likes
    tweet = (textContent, dateTime, replies, retweets, likes)
    
    #Tweet will be returned.
    return tweet
```

<h4>Part 4(ii): Implementing the function to webscrape Elon Musk's Profile page</h4>
<p>At this stage, to implement the code to webscrape Elon Musk's Profile page, what I did was to:
    <ol>
        <li>Identify tweets in a single page - This is done by using the find_elements function and finding the XPATH of the tweet.</li>
        <li>Using a for loop to loop through the results of the elements found, passing in that single tweet as input to the get_tweet_contents function and appending the returned data into the list ElonMuskTweets.</li>
        <li>Once all the tweets in a page is successfully scraped, the webscraper will scroll down to load more tweets.</li>
        <li>Once the webscraper is unable to scroll down more than or equal to 3 times, it will stop the process.</li>
    </ol>
</p>

```python
#Retrieve all tweets by elon musk
ElonMuskTweets = []

tweet_ids = set()

#Get last scroll position 
previous_position = driver.execute_script("return window.pageYOffset;")
scrolling = True

while scrolling:
    #Retrieve tweets from 
    page_Tweets = driver.find_elements(By.XPATH, '//article[@data-testid="tweet"]')

    #Loop through all tweets in a twitter page
    for singleTweet in page_Tweets:
        #Retrieve all the contents in a single tweet using the function which we made above
        oneTweet = get_tweet_contents(singleTweet)
        
        #If there is a tweet, append said data. Basically adding data into a list
        if(oneTweet):
            ElonMuskTweets.append(oneTweet)
            
    scrolling_attempts = 0
    while True: 
        #Scrolling
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        sleep(1)
        #Get current position
        curr_position = driver.execute_script("return window.pageYOffset;")
        if previous_position == curr_position:
            scrolling_attempts += 1

            #end of scroll
            if scrolling_attempts >= 3:
                scrolling = False
                break
            else:
                #Try to scroll again
                sleep(2)
        else:
            #Set last scroll position to current position and break
            previous_position = curr_position
            break
```

<h3>Processing the CSV data and filtering out the Content words</h3>
<p>At this stage, we will be filtering out the stop words. This is done by tokenizing the words in the 'Content' columns in the csv file and storing it in the words list. This will be done row by row. After completing this process, all the stored words in the list will be filtered, leaving the necessary words in the words list. I will be using the following items from nltk:
    <ol>
        <li>punkt - find how different sentences can be related to each other, combining them together</li>
        <li>stopwords - remove stopwords </li>
        <li>wordnet</li>
    </ol>
Using the HanselAndGretelExample.ipy from week 5 coursera, I managed to implement a way to remove stopwords and analyze or remove common words in Elon Musk tweets.

<h3>Removing illegal values / values which is redundant</h3>
<img src="photos/EmptyContentCsv.png">
<img src="photos/ElonMuskTweetNoText.png">
<p>One issue that I faced is that certain tweets by Elon Musk has no text. It will result in scraping a tweet with empty content, An example is where Elon Musk tweets a single emoji 'top'. As a result, the content in the csv file will be blank. The two photos above with a red box highlights the tweet and the empty content in the csv file. To circumvent this problem, I had to implement an if statement in the for loop which access the content rows. If the content type checked is a 'str' (string), the tokenization will occur. If it is not a 'str' (string), tokenization will not happen and the row will be skipped.</p>


```python
#Create a list of words
words = []

# Read the csv file from elon musk and save to data
data = pd.read_csv('elonMuskTweets.csv', encoding = 'utf-8')

#Get all the indexes in data
index = data.index

#Get the content in the data row 'Content' by using a for loop
for row in data['Content']:
    #Processing data, if the type value of the row accesses is not str, tokenization will not be executed.
    if(type(row) == str): 
        #Tokenize a single row in the 'Content' and insert the tokenized words into words.
        words = words + nltk.word_tokenize(row)
    
# Contains frequency analysis of the words in the csv file
analysis = nltk.FreqDist(words)

#Remove stop words
filter_words = [word for word in words if not word in stopwords.words()]

# Create dictionary of objects
filter_words = dict([(m, n) for m,n in analysis.items() if len(m) > 4])
wcloud = WordCloud(background_color = "white",width=2000, height=2000).generate_from_frequencies(filter_words)
```

<h3>Plotting the graph to display wordcloud</h3>

```python
#Make the figure larger
plt.figure( figsize=(20,15) )

#show the words in a plot
plt.imshow(wcloud, interpolation = "bilinear")

#plot graph without axis
plt.axis("off")
(-1.5, 200, 100, -2.5)

#Show the wordcloud
plt.show()
```

<h2>Part 1: Retrieving dates where Elon Musk mentions Tesla / Twitter in his tweet</h2>
<h3>Creating a Function to retrieve dates Elon Musk mentions Tesla / Twitter</h3>
<p>As we are retrieving two different keywords Elon Musk mentions in his tweets, the most feasible plan is to create a function which takes the keyword as input and it will output a list of dates where Elon Musk mentions the keywords. To retrieve the dates, I would need to iterate through the csv file containing Elon Musk tweets row by row. At each for loop iteration, the for loop will:
    <ol>
        <li>Check if the keyword is mentioned in the tweet contents, if true, the date/time string for that row will be copied to a local variable</li>
        <li>The variable containing the date/time will be converted into from string type to dateTime type, then removing the time to date only</li>
        <li>The variable will then be appended to a list storing the dates of the tweets mentioning the keyword</li>
    </ol>
</p>

<h3>Data Cleaning</h3>
<h4>Date and time representation in elonMuskTweets.csv</h4>
<img src="photos/DateTimeCsv.png" alt="Image Displaying the date and time in Elon Musk tweets csv">
<h4>Date representation in TeslaStockPrices.csv</h4>

<p>One of the problems I had faced when creating the function is that the csv containing Elon Musk Tweets has the date and time together in its column which can be seen in the photo above while the other csv file (Stock Market Price of Tesla) contains only the date. Furthermore, both csv files which contains the dates are in the string type instead of the date/time type. Hence, it will be hard to compare the dates between the two csv files. Thus, I will need to convert the date and time of the tweet into the date only. As I am removing information which is the time of the tweet, I will not be editing the csv file directly as I may need the time of tweets for future analysis. I will be copying the information over from the csv and edit the copied data instead.</p>


```python
#Import parser and datetime libraries
from datetime import datetime  
from dateutil import parser

#Function to retrieve dates where keyword is mentioned. Returns list of dates
def datesKeywordMentioned(keyword):
    #Dataframe to store the mentioned dates in the dataframe
    mentionedDates = []

    # Read the csv file from elon musk and save to data
    data = pd.read_csv('elonMuskTweets.csv', encoding = 'utf-8')


    #Get all the indexes in data
    index = data.index

    #Iterate through the rows in the csv rows
    for index, row in data.iterrows(): 

        if(type(row['Content']) == str):

            #If there is mention of Content word in tweet content
            if(keyword in row['Content']):

                #Data cleaning up
                #Convert from string date time into dateTime type
                date_time = parser.parse(row['dateTime'])

                #Concert date time into date only
                date = date_time.date()

                #Insert tweeted dates into mentionedDates dataframe
                mentionedDates.append(date)

                print('Dates mentioned:',date)

    #Check if the dates has been successfully saved
    print('Number of tweets Elon Musk mentions', keyword,':',len(mentionedDates))
    #Return the list
    return mentionedDates
```

```python
#Retrieve dates where Tesla is mentioned
datesMentionedTesla = datesKeywordMentioned("Tesla")

#Retrieve dates where Twitter is mentioned
datesMentionedTwitter = datesKeywordMentioned("Twitter")
```

<h2>Part 2: Finding Stock Market data on dates where Elon Musk mentions Tesla</h2>
<h3>Creating a function to calculate price differences of a particular companies stock</h3>
<p>The input arguments for this particular function called calculatePriceDifferences will take in 2 values. They are dateList and StockMarketFile. dateList is the list of dates which Elon Musk mentions a certain company. In our case, they are the companies Twitter and Tesla. The StockMarketFile are just csv files containing historical data on a particular company's stock prices. How this function will function is described in a series of steps specified below.</p>
<h4>Steps taken in the function:</h4>
    <ol>
        <li>Read the csv file on past historical data on a company Stock Price called: companyA.csv.</li>
        <li>Iterate through the rows in that csv file, checking for matches on the days Elon Musk mentions companyA and days on the historical stock market data on companyA.</li>
        <li>On each match, the price difference between the columns Open and Close for that row (that particular date in the stock market) will be calculated to determine if there is a general increase in price or decrease in price for company A's stock.</li>
    </ol>

<h3>Problems I had faced</h3>
    <ol>
        <li>The dates in the csv taken from Yahoo Finance is in string type, I had to convert it into date/time type and then change it into date only.</li>
        <li>When calculating the price differences for Tesla, there are 17 times where elon musk mentioned Tesla in his tweets but only 12 matches were made. This is because the other 5 non-matches are on days which is on the weekends or holidays. (e.g: 5th July 2021 - Independence day) On those days, the stock market are not open. It took me a while to figure this out.</li>
    </ol>

```python
def calculatePriceDifferences(dateList, StockMarketFile):

    #Counter to see dates matches between stock market and dates where elon musk mentions 'Keyword'
    noOfMatches = 0

    # Read the csv file for a particular stockmarket and save to dataStock
    dataStock = pd.read_csv(StockMarketFile, encoding = 'utf-8')

    #Get all the indexes in data
    index = dataStock.index

    #List to store the price differences
    priceDifferences = []

    #For loop to iterate through the rows in the csv rows
    for index, row in dataStock.iterrows():
        #Loop through the mentionedDates

        #Convert from string date time into dateTime type
        dateStockPrices = parser.parse(row['Date']).date()

        #For loop to iterate through the list of dates 
        for x in dateList:
           #Checks if dates in list matches a day in the stock market historical price
            if(x == dateStockPrices):
                #Get highest price for that day
                highPrice = row['High']
                #Get lowest price for that day
                lowPrice = row['Low']

                #Calculate the difference between the highest and lowest price difference for that day
                highLowDiff = highPrice - lowPrice
                
                #Insert the result into the dataframe
                priceDifferences.append(highLowDiff)

                print('Date:',dateStockPrices,', High and Low price difference:',highLowDiff)

                #Increment the number of matches by one
                noOfMatches += 1

    print('\nTotal number of matched dates:',noOfMatches)
    return priceDifferences
```

```python
#check the price differences of Tesla stock when Elon Musk mentions Tesla
print("Tesla price differences:")
TeslaPriceDifferences = calculatePriceDifferences(datesMentionedTesla, 'TeslaStockPrices.csv')

#Check the price differences of Twitter stock when Elon Musk mentions twitter
print("\nTwitter price differences:")
TwitterPriceDifferences = calculatePriceDifferences(datesMentionedTwitter, 'TwitterStockPrices.csv')
```

<h2>Part 3: Finding the average Close and Open stock price difference</h2>
<h3>Creating a function to find the average close and open differences in price</h3>
<p>We will be calculating the average Close and Open stock price difference to see if Elon Musk's tweets actually makes that price difference be higher than average. As we are calculating more than one companies stock prices, we will need to create a function where we will take the companies past stock market prices in a csv file format. The function will perform the following steps:
<ol>
    <li>Loop through all of the rows in the csv file</li>
    <li>Calculate the price difference in each rows</li>
    <li>Sum up the price difference in all the rows</li>
    <li>Average the sum out</li>
    <li>Return the average sum</li>
</ol>
After performing the steps,using the average sum returned by this function, I will be able to determine if the difference in prices of Tesla when the stock market opens and closes is higher than average when Elon Musk mentions Tesla / Twitter in his tweets for that day.
</p>

```python
def calculatePriceAverage(StockMarketFile):
    # Read the TeslaStockPrices csv file and save to dataStock
    dataStock = pd.read_csv(StockMarketFile, encoding = 'utf-8')

    #Get all the indexes in data
    index = dataStock.index

    #Count total number of days stock market is open
    counter = 0

    #Stores total price difference (Open / Close)
    sumPriceDiff = 0

    #Stores result
    avePriceDiff = 0

    #For loop to iterate through the rows in the csv rows
    for index, row in dataStock.iterrows():
        #Get Tesla open price for that day
        highPrice = row['High']
        #Get Tesla close price for that day
        lowPrice = row['Low']

        #Calculate the price difference when Stock Market Opens and when Stock Market Closes
        highLowDiff = highPrice - lowPrice

        #Collate the total priceDifference
        sumPriceDiff += highLowDiff

        #Increment the number of matches by one
        counter += 1


    avePriceDiff = sumPriceDiff / counter
    print('The average price difference of Highest and Lowest price for',StockMarketFile,'is:',avePriceDiff)
    return avePriceDiff

            
```

```python
teslaAveragePrice = calculatePriceAverage('TeslaStockPrices.csv')
twitterAveragePrice = calculatePriceAverage('TwitterStockPrices.csv')
```

<h2>Part 4: Comparing the price differences on the days Elon Musk Mentions twitter with the average price difference of a Twitter / Tesla</h2>
<p>With the average Price Difference of Twitter / Tesla calculated, I am now able to compare the list containing the price differences of Twitter / Tesla on the days Elon Musk mentions about them. To perform the comparison, I will create a function as we are comparing more than one company. The function will perform the following steps:
    <ol>
        <li>Use a for loop to loop through the list containing the price differences between the lowest and highest price on the day Elon Musk mentions about Twitter / Tesla.</li>
        <li>On each iteration, a comparison will be made if the price difference on the days Elon Musk Mentions Twitter / Tesla is higher than the average price differences.</li>
        <li>If the price difference is higher than average, a counter named noTimesHigherAve will be incremented by one</li>
        <li>One the for loop fully executes, the counter will be divided by the length of the list to determine the percentage of days where the price change will be higher than average</li>
    </ol>
    This percentage calculated will represent the chances of the price difference of Twitter / Tesla stocks that day being higher than average when Elon Musk mentions Twitter / Tesla in his tweets.
</p>

```python
def calculatePercentage(averagePrice, priceDifferences):
    #Counter to store the number of days price difference is higher than average
    noTimesHigherAve = 0

    #Loop through the calculated priceDifferences
    for x in priceDifferences:
        #If calculated priceDifferences is higher than average
        if (x > averagePrice):
            #Increment the number of times price is higher than average
            noTimesHigherAve += 1

    #Calculate the percentage
    percentage = noTimesHigherAve / len(priceDifferences) * 100

    return percentage
    
```

```python
print('Chances of Tesla Stock prices to be higher than average when Elon Musk mentions Tesla:',calculatePercentage(teslaAveragePrice, TeslaPriceDifferences),'%')
print('Chances of Twitter Stock prices to be higher than average when Elon Musk mentions Twitter:',calculatePercentage(twitterAveragePrice, TwitterPriceDifferences), '%')
```

<h1>Sentiment Analysis on Elon Musk tweets</h1>
<p>The next thing I will be doing is to perform a sentiment analysis on his tweets. To do this I will go through the contents of the tweets by Elon Musk in the csv file line by line. In each line the following steps will happen:
<ol>
    <li>Create 3 lists containing the positive / negative / neutral tweets for future use</li>
    <li>Analyze the tweet content in question and store the positive / negative / neutral score as individual values</li>
    <li>The values will be compared with each other and they will be classified and inserted into one of the three list. A counter for positive / negative / neutral tweets will be incremented accordingly.</li>
</ol>
Once done, the incremented values for positive / negative / neutral values will be used to create a pie chart to display how positive / negative / neutral Elon Musk tweets are.
</p>

```python
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import math
```

```python
# Read the csv file from elon musk and save to data
data = pd.read_csv('elonMuskTweets.csv', encoding = 'utf-8')

#Integers storing positive, negative and neutral Numbers of tweets
positiveNumbers = 0
negativeNumbers = 0
neutralNumbers = 0
totalNumbertweets = 0

#Lists storing positive, negative and neutral
positiveList = []
negativeList = []
neutralList = []

for row in data['Content']:
    if(type(row) == str):
    
        #Calculate the analyzed score of each tweet by elon musk
        AnalyzedScore = SentimentIntensityAnalyzer().polarity_scores(row)
    
        #Positiveness of tweet
        positiveScore = AnalyzedScore['pos']
    
        #Negativeness of tweet
        negativeScore = AnalyzedScore['neg']
    
        #Neutralness of tweet
        neutralScore = AnalyzedScore['neu']
    
    
        #Store the tweets depending if it is negative, neutral or positive
        #If tweet is positive
        if (positiveScore > negativeScore):

            #Store positive tweet in the positive list
            positiveList.append(row)

            #Increment the number of positive tweets sent by elon by one
            positiveNumbers += 1

        #If tweet is negative
        elif (positiveScore < negativeScore):
            #Store negative tweet in the negative list
            negativeList.append(row)

            #Increment the number of negative tweets send by elon by one
            negativeNumbers += 1

        #If tweet is neutral
        elif (positiveScore == negativeScore):
            #Store neutral tweet in the neutral list
            neutralList.append(row)

            #Increment the number of neutral tweets sent by elon by one
            neutralNumbers += 1
        
#Finalize the total number of tweets analyzed
totalNumbertweets = neutralNumbers + negativeNumbers + positiveNumbers



#Calculating the percentage of positive, negative or neutral tweets
#Positive tweets percentage
positivePercent = math.ceil(positiveNumbers / totalNumbertweets * 100)

#Negative tweets percentage
negativePercent = math.ceil(negativeNumbers / totalNumbertweets * 100)

#Neutral tweets percentage
neutralPercent = math.floor(neutralNumbers / totalNumbertweets * 100)

print('Total percentage:',positivePercent + negativePercent + neutralPercent)
```

```python
#Labels for pie chart
labels = ['Positive tweets: '+str(positivePercent)+'%', 'Negative tweets: '+str(negativePercent)+'%', 'Neutral tweets: '+str(neutralPercent)+'%']

#Pie chart sizes
sizes = [positivePercent, negativePercent, neutralPercent]
#Set the colors for positivePercent, negativePercent, neutralPercent respectively
colors = ['green', 'red', 'blue']
patches, texts = plt.pie(sizes, colors = colors, startangle = 90)
plt.style.use('default')
plt.legend(labels)
plt.title('Elon Musks percentage of positive, negative and neutral tweets')
plt.show()

```

<h1>Further analysis on Elon Musk tweets on Twitter / Tesla</h1>
<p>When I was performing sentiment analysis on Elon Musk tweets, decided to categorize the positive / negative / neutral tweets into individual lists which will be used in this portion of our report. The aim of this analysis is to determine if Elon Musk mentioning a company positively / negatively will affect its stock prices in a bad / good way. To do this, I will implement a function which will:
<ol>
    <li>Take in the keyword and the list of tweets as the function inputs</li>
    <li>Loop through the list of tweets and determine if the keyword is mentioned in any if the tweets</li>
    <li>If the keyword is mentioned, a counter will increment by one</li>
    <li>The counter will be returned at the end of the loop</li>
</ol>
This function will be used to find out the number of times Elon musk mentions 'Tesla' and 'Twitter' positively / negatively / neutrally.
</p>

```python
#Function to count number of times Elon Musk mentions keyword in his tweets
def countMentionsTweets(keyword, tweetList):
    #Dataframe to store the mentioned dates in the dataframe
    mentionedDates = []
    
    #Counter to increment the number of times elon musk mentions the company
    counter = 0

    #Iterate through the rows in the csv rows
    for x in tweetList: 
        #If there is mention of word in tweet content
        if(keyword in x):
            counter += 1
                
    #Return the list
    return counter
```
